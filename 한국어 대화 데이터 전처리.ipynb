# 한국어 연속성 대화

```python
# 한국어_연속적_대화_데이터셋.xlsx
data_yun_1 = pd.read_excel('/home/aiffel-dj20/Downloads/데이터셋/한국어_연속적_대화_데이터셋.xlsx')
data_yun_1.head()
```

```python
data_yun_1.columns
```

```python
data_yun_one = data_dan_1['Unnamed: 1']
data_yun_one
```

```python
data_yun_one = data_dan_one.dropna()
data_yun_one.isnull().sum()
```

```python
data_yun_one
```

```python
data_yun_one = data_yun_one.reset_index()
data_yun_one
```

```python
Q = data_yun_one[::2]
A = data_yun_one[1::2]
```

```python
Q.name = 'Q'
A.name = 'A'
```

```python
Q.head()
```

```python
A.head()
```

```python
Q = Q.reset_index(drop=True)
```

```python
A = A.reset_index(drop=True)
```

```python
Q = Q.dropna()
A = A.dropna()
```

```python
df = pd.merge(Q,A,how='outer', left_index=True, right_index=True)
```

```python
df.head(50)
```

```python
df.to_csv('한국어_연속_대화.csv')
```
